## Q1. What is Flask Framework? What are the advantages of Flask Framework?
### Asignement:-
### Flask is a lightweight web framework written in Python. It is designed to make it easy to build web applications and APIs, with a simple and intuitive API and minimal setup required. Flask is often used for small to medium-sized projects, where simplicity and flexibility are important.

### Advantages of Flask Framework:

### Lightweight: Flask is a lightweight framework that doesn't require much setup or configuration, making it easy to get started quickly.

### Flexibility: Flask is highly flexible and allows developers to use their preferred tools and libraries for things like databases, templates, and authentication.

### Simplicity: Flask has a simple and intuitive API that makes it easy to build web applications and APIs without getting bogged down in complex details.

### Extensibility: Flask has a large number of third-party extensions available, which can be easily added to enhance the functionality of the framework.

### Compatibility: Flask is compatible with many other Python libraries and tools, making it easy to integrate with existing systems.

### Built-in Development Server: Flask comes with a built-in development server, which makes it easy to test and debug applications during development.

### Large Community: Flask has a large and active community of developers, which means that there are many resources and tutorials available to help developers get started and solve problems.

## Q3. What is App routing in Flask? Why do we use app routes?
### Answer:-
### In Flask, app routing refers to the process of mapping URLs or routes to specific functions in your application. The routes specify what URL a user should visit to trigger a specific function within your application.

### App routes are used in Flask to define the various endpoints that users can visit in your web application. These endpoints can be thought of as the different pages of your application. When a user visits a particular URL or route, Flask will look up the corresponding function and execute it, returning the output to the user.

### For example, you can use app routes to define a route for your home page, a route for your contact page, and a route for your blog page. This way, when a user visits your website and clicks on a link, Flask knows exactly which function to call to serve up the corresponding page.
